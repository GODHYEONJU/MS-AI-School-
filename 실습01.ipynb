```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 데이터셋 불러오기
data = pd.read_csv("../BlackFriday.csv")
print(data)
```

```python
# User_ID별 구매 총 금액 계산
user_total_spent = data.groupby('User_ID')['Purchase'].sum()
```

```python
# 히스토그램 그리기
plt.hist(user_total_spent, bins=50)
plt.title('Histogram of Total Purchase Amout by User')
plt.xlabel('Total Purchase Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
# 밀도 그래프 그리기
plt.figure()
user_total_spent.plot.kde()
plt.title('Density Plot of Total Purchase Amount by User')
plt.xlabel('Total Purchase Amount')
plt.show()
```
