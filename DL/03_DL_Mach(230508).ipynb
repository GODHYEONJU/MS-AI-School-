```python
#신경망 기초

import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1차 함수
#변수 생성
def linear_function(x):
  a=0.5
  b=2

  return a*x+b
```

```python
print(linear_function(5))
```

```python
x = np.arange(-5.5,0.1)
y = linear_function(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Function')
```

```python
# 2차 함수
def quadratic_function(x):
  a = 1
  b = -1
  c = -2

  return a*x**2 + b*x + c
```

```python
print(quadratic_function(2))
```

```python
x = np.arange(-5,5,0.1)
y = quadratic_function(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic Function')
```

```python

#3차 함수
def cubic_function(x):
  a = 4
  b = 0
  c = -1
  d = -8

  return a*x**3 + b*x**2 + c*x + d
```

```python
x = np.arange(-5,5,0.1)
y = cubic_function(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cublic Function')
```

```python
# 논리 게이트

# AND
def AND(a,b):
  input = np.array([a,b])
  weigths = np.array([0.4,0.4])
  bias = - 0.6
  value = np.sum(input * weigths) + bias

  if value <= 0:
    return 0
  else:
    return 1
```

```python
print(AND(0,0))
print(AND(0,1))
print(AND(1,0))
print(AND(1,1))
```

```python
x1 = np.arange(-2,2,0.01)
x2 = np.arange(-2,2,0.01)
bias = -0.6

y = (-0.4*x1 - bias)/0.4

plt.axvline(x=0)
plt.axhline(y=0)

plt.plot(x1, y, 'r--')
plt.scatter(0,0,color='orange', marker='o', s=150)
plt.scatter(0,1,color='orange', marker='o', s=150)
plt.scatter(1,0,color='orange', marker='o', s=150)
plt.scatter(1,1,color='black', marker='^', s=150)

plt.xlim(-0.5,1.5)
plt.ylim(-0.5,1.5)
plt.grid()
```

```python
# OR
def OR(a,b):
  input = np.array([a,b])
  weigths = np.array([0.4,0.4])
  bias = - 0.3
  value = np.sum(input * weigths) + bias

  if value <= 0:
    return 0
  else:
    return 1
```

```python
print(OR(0,0))
print(OR(0,1))
print(OR(1,0))
print(OR(1,1))
```

```python
# NAND
def NAND(a,b):
  input = np.array([a,b])
  weigths = np.array([-0.5,-0.5])
  bias = 0.7
  value = np.sum(input * weigths) + bias

  if value <= 0:
    return 0
  else:
    return 1
```

```python
print(NAND(0,0))
print(NAND(0,1))
print(NAND(1,0))
print(NAND(1,1))
```

```python
x1 = np.arange(-2,2,0.01)
x2 = np.arange(-2,2,0.01)
bias = 0.7

y = (0.5*x1 - bias)/-0.5

plt.axvline(x=0)
plt.axhline(y=0)

plt.plot(x1, y, 'r--')
plt.scatter(0,0,color='orange', marker='o', s=150)
plt.scatter(0,1,color='orange', marker='o', s=150)
plt.scatter(1,0,color='orange', marker='o', s=150)
plt.scatter(1,1,color='black', marker='^', s=150)

plt.xlim(-0.5,1.5)
plt.ylim(-0.5,1.5)
plt.grid()
plt.show()
```

```python
# XOR
def XOR(x1, x2):
  s1 = NAND(x1, x2)
  s2 = OR(x1, x2)
  y = AND(s1, s2)

  return y
```

```python
print(XOR(0,0))
print(XOR(0,1))
print(XOR(1,0))
print(XOR(1,1))
```
